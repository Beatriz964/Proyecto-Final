### Clásico
$\textbf{Espacio muestral:}$

$ \Omega = \{\text{todos los posibles resultados de un experimento}\} $

$\textbf{Probabilidad Clásica:}$

$P(A) = \frac{|A|}{|\Omega|}$

$\textbf{Probabilidad Condicional:}$

$P(A \mid B) = \frac{P(A \cap B)}{P(B)}, \quad \text{si } P(B) > 0$

Es la probabilidad de que ocurra $ A $ sabemos que ocurrió $ B $. 
Se reduce el espacio muestral a $ B $.

$\textbf{Independencia:}$

$ P(A \cap B) = P(A)P(B) $

$\textbf{Probabilidad del Complemento:}$

$ P(A^c) = 1 - P(A)$

Como $ A \cup A^c = S $ y $ A \cap A^c = \varnothing $

$P(S) = P(A) + P(A^c) \Rightarrow P(A^c) = 1 - P(A)$

$\textbf{Probabilidad de la Unión}$

$ P(A \cup B) = P(A) + P(B) - P(A \cap B) $

$ |A \cup B| = |A| + |B| - |A \cap B| \Rightarrow \frac{|A \cup B|}{|S|} = \frac{|A|}{|S|} + \frac{|B|}{|S|} - \frac{|A \cap B|}{|S|} $

$\textbf{Leyes fundamentales de la probabilidad con conjuntos}$

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

Si $ A \cap B = \emptyset $, entonces: $P(A \cup B) = P(A) + P(B)$


### Geométrico

Se utiliza cuando los resultados posibles de un experimento corresponden a puntos dentro de una región del espacio y se mide una longitud, área y/o volumen. 
$$P(A) = \frac{\mu(A)}{\mu(S)}$$

donde $ \mu $ es una **medida** (longitud, área, etc.), y $ A \subseteq S $.


```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(0, 60)
ax.set_ylim(0, 60)
ax.set_xlabel("Llegada A (min)")
ax.set_ylabel("Llegada B (min)")
# Espera A=10 min, B=20 min"

x = np.linspace(0, 60, 1000)
# Región donde sí se encuentran
ax.fill_between(x, np.maximum(0, x - 20), np.minimum(60, x + 10), color="lightgreen", label="Se encuentran")

# Líneas guía
ax.plot(x, x + 10, 'r--', label="Y = X + 10")
ax.plot(x, x - 20, 'b--', label="Y = X - 20")
ax.legend()
ax.grid(True)
plt.show()
```

### Frecuentista
La probabilidad frecuentista se basa en la **frecuencia relativa** observada de un evento en múltiples repeticiones del mismo experimento.

Sea $ A $ un evento de interés, y $ n(A) $ el número de veces que ocurre después de $ n $ repeticiones, entonces:

$$
P(A) \approx \frac{n(A)}{n}
$$

Cuando $ n \to \infty $, esta proporción converge (por la ley de los grandes números):

$$
P(A) = \lim_{n \to \infty} \frac{n(A)}{n}
$$

```python
#----------------------------------------------
#              EJEMPLO DEL VOLADO
#-----------------------------------------------
import random

def simular_volado(n=100):
    resultados = [random.choice(['cara', 'cruz']) for _ in range(n)]
    caras = resultados.count('cara')
    return caras / n

frecuencias = [simular_volado(n=100) for _ in range(100)]

import matplotlib.pyplot as plt
plt.hist(frecuencias, bins=20, edgecolor='black')
plt.xlabel('Frecuencia relativa')
plt.ylabel('Frecuencia')
plt.axvline(0.5, color='red', linestyle='dashed')
plt.grid(True)
plt.show()
```
